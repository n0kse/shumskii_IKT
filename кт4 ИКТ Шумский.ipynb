1. Загрузить фрейм данных 'demo26.xlsx' с сайта https://github.com/junaart/ForStudents/tree/master/Big_Data/Seminar_3

```python
import pandas as pd
import numpy as np
url = "https://github.com/junaart/ForStudents/raw/master/Big_Data/Seminar_3/demo26.xlsx"
data = pd.read_excel(url, sheet_name='ОПЖ')
data
```

```python
2. Создать фрейм данных, транспонируя исходный фрейм данных, присвоив имена столбцов 'FIRST', 'SECOND', 'THIRD', 'FOURTH', 'FIFTH'.
```

```python
dd=data[[1990, 1995, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]]
dd.index=data['Год']
display(dd)
```

```python
d1=dd.T
d1
```

```python
d1=d1.rename(columns={"Все население": "FIRST", "Все мужчины" : "SECOND","Все женщины" : "THIRD", "Все городское население": "FOURTH", "Все городское население мужчины": "FIFTH"})
dM=d1[["FIRST", "SECOND", "THIRD", "FOURTH", "FIFTH"]]
dM
```

3. Отобрать года, для которых значение во втором столбце больше среднего значения этого столбца

```python
dM1 = dM[dM['SECOND'] > dM['SECOND'].mean()]
dM1
```

4. Вывести на экран все строки фрейма за четные года.

```python
dM2 = dM[dM.index % 2 == 0]
dM2
```

5. Построить график по оси x - года, по оси y значения третьего столбца.

```python
import matplotlib.pyplot as plt
plt.plot(dM.index, dM['THIRD'], marker='o', linestyle='-', color='b')
plt.title('График значений третьего столбца по годам')
plt.xlabel('Год')
plt.ylabel('THIRD')
```

6. Добавить к фрейму новый столбец категориальных данных по принципу: если значение третьего столбца в строке больше медианного значения этого столбца и значение четвертого столбца этой строки больше медианного значения этого столбца, то в новый столбец пишем 'good', в противном случае пишем 'bad'.

```python
dM['Категория'] = dM.apply(lambda row: 'good' if (row['THIRD'] > dM['THIRD'].median()) and (row['FOURTH'] > dM['FOURTH'].median()) else 'bad', axis=1)
dM
```
