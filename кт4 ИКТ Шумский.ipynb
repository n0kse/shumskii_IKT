2. Создать фрейм данных, транспонируя исходный фрейм данных, присвоив имена столбцов 'FIRST', 'SECOND', 'THIRD', 'FOURTH', 'FIFTH'.

```python
data.columns.values
```

```python
data=data.rename(columns={"Год": "FIRST", 1990: "SECOND", 1995: "THIRD", 2000: "FOURTH", 2001: "FIFTH"})
data
```

3. Отобрать года, для которых значение во втором столбце больше среднего значения этого столбца.

```python
data[data['SECOND'] > data['SECOND'].mean()][['FIRST','SECOND', 'THIRD', 'FOURTH', 'FIFTH']]
```

3. Построить график по оси x - года, по оси y значения третьего столбца.


```python
import matplotlib.pyplot as plt
plt.plot(data['FIRST'], data['THIRD'], marker='o', linestyle='-', color='b')
plt.title('График значений третьего столбца по годам')
plt.xlabel('FIRST')
plt.ylabel('THIRD')
```

```python
median_col3 = data['SECOND'].median()
median_col4 = data['THIRD'].median()
data['Категория'] = data.apply(lambda row: 'good' if (row['SECOND'] > median_col3) and (row['THIRD'] > median_col4) else 'bad', axis=1)
```
